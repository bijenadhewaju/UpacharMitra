## 1: Load Dataset

```python
from datasets import load_dataset, concatenate_datasets, get_dataset_config_names, DatasetDict
from sklearn.model_selection import train_test_split

# Step 1: Get all medical specialty configs
configs = get_dataset_config_names("HPAI-BSC/medical-specialities")

# Step 2: Load and label a subset from each specialty
all_datasets = []

for config in configs:
    try:
        print(f"Loading {config}...")
        dataset = load_dataset(
            "HPAI-BSC/medical-specialities",
            config,
            split="test[:200]",  # change to "test" for full, or increase slice
            trust_remote_code=True
        )
        # Add label field
        dataset = dataset.map(lambda x: {"label": config})
        all_datasets.append(dataset)
    except Exception as e:
        print(f"Failed to load {config}: {e}")

# Step 3: Concatenate datasets
if all_datasets:
    full_dataset = concatenate_datasets(all_datasets)
    print(f"\n Combined dataset loaded with {len(full_dataset)} examples.")

    # Step 4: Split into train and test (80/20 split)
    train_test = full_dataset.train_test_split(test_size=0.2, seed=42)
    dataset_dict = DatasetDict({
        'train': train_test['train'],
        'test': train_test['test']
    })

    # Sample view
    print("\n Sample example:")
    print(dataset_dict['train'][0])

else:
    print("No datasets loaded.")
```

#### Preparing the final dataset for training

```python
from datasets import Dataset
from sklearn.preprocessing import LabelEncoder

# Encode the labels
label_encoder = LabelEncoder()
filtered_df["label"] = label_encoder.fit_transform(filtered_df["medical_field"])

# Save mappings for inference later
label2id = {label: int(idx) for label, idx in zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_))}
id2label = {idx: label for label, idx in label2id.items()}

# Convert to Hugging Face Dataset
hf_dataset = Dataset.from_pandas(filtered_df)
```

```python
from datasets import DatasetDict

dataset_dict = hf_dataset.train_test_split(test_size=0.2, seed=42)
```

```python
from transformers import AutoTokenizer

# Load tokenizer
tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased")

# Tokenization function
def tokenize_function(example):
    return tokenizer(example["question"], padding="max_length", truncation=True)

# Apply tokenization
tokenized_dataset = hf_dataset.map(tokenize_function, batched=True)

# Remove original text and keep label
tokenized_dataset = tokenized_dataset.remove_columns(["question", "medical_field"])
tokenized_dataset.set_format("torch")
```

```python
split_dataset = tokenized_dataset.train_test_split(test_size=0.2)
train_dataset = split_dataset["train"]
eval_dataset = split_dataset["test"]
```

```python
from transformers import AutoModelForSequenceClassification

num_labels = len(label2id)

model = AutoModelForSequenceClassification.from_pretrained(
    "distilbert-base-uncased",
    num_labels=num_labels,
    id2label=id2label,
    label2id=label2id
)
```

```python
from transformers import pipeline

# Option A: Use pipeline (easy for common tasks)
classifier = pipeline("text-classification", model=model, tokenizer=tokenizer)

samples = [
    "i have sore throat"
]

predictions = classifier(samples)

for i, pred in enumerate(predictions):
    print(f"Input: {samples[i]}")
    print(f"Prediction: {pred}")
```

```python
import torch

model.eval()

# Move the model to the appropriate device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

inputs = tokenizer("Your test sentence here", return_tensors="pt")
# Move the inputs to the same device as the model
inputs = {key: val.to(device) for key, val in inputs.items()}


with torch.no_grad():
    outputs = model(**inputs)

# For classification, outputs.logits usually contain raw scores

logits = outputs.logits
predicted_class_id = logits.argmax(dim=-1).item()

print(f"Predicted class id: {predicted_class_id}")
predicted_label = id2label[predicted_class_id]
print(f"Predicted label: {predicted_label}")
```
